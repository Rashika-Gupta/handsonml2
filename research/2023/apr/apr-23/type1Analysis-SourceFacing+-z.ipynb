## tot counts Type 1

```python
def nCountsDet (df, tDet1, tDet2, det1, Ecut):
    df = df[df[det1] > Ecut]
    df = df[df[tDet1] < df[tDet2]]
    n = df.shape[0] # total count
    return n
```

## avgX vs Y

```python
def avgXavgY(df, X, Y, tFirst, tLast, nBins, minRange, maxRange, Ecut, detTrig,) :
    if Ecut != 0 :
        df = df[df[detTrig] > Ecut]
        df = df[df[tFirst]<df[tLast]]
    df['etof'] = df[tLast]-df[tFirst]
    mean_stat  = scipy.stats.binned_statistic(x = df[X], values = df[Y], statistic='mean', bins = nBins, range = (minRange,maxRange))
    count_stat = scipy.stats.binned_statistic(x = df[X], values = df[Y], statistic='count', bins = nBins, range = (minRange,maxRange))
    xEdge = mean_stat.bin_edges[1:]
    yMean = mean_stat.statistic
    yCount = count_stat.statistic
    yStd = yMean/np.sqrt(yCount)
    return xEdge, yMean, yStd, yCount
```
